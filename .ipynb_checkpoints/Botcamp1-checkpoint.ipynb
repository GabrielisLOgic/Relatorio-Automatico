```python
import pandas as pd
import datetime
import yfinance as yf
from matplotlib import pyplot as plt
import mplcyberpunk
import smtplib
from email.message import EmailMessage
```

```python
ativos = ["^BVSP", "BRL=X", "^GSPC", "BTC=F"]

hoje = datetime.datetime.now()
umAno = hoje - datetime.timedelta(days = 365)

dadosMercado = yf.download(ativos, umAno, hoje)

display(dadosMercado)
```

```python
dadosFechamento = dadosMercado['Adj Close']
dadosFechamento.columns = ['Dolar', 'Bitcoin', 'Ibovespa', 'S&P500']
dadosFechamento = dadosFechamento.dropna()

dadosFechamento
```

```python
dadosFechamentoMensal = dadosFechamento.resample("M").last()
dadosFechamentoAnual = dadosFechamento.resample("Y").last()

dadosFechamentoAnual
```

```python
retornoAnual = dadosFechamentoAnual.pct_change().dropna()
retornoMensal = dadosFechamentoMensal.pct_change().dropna()
retornoDiario = dadosFechamento.pct_change().dropna()



retornoAnual
```

```python
AnualDolar = retornoAnual.iloc[-1, 0]
MensalDolar = retornoMensal.iloc[-1, 0]
DiarioDolar = retornoDiario.iloc[-1, 0]

AnualBitcoin = retornoAnual.iloc[-1, 1]
MensalBitcoin = retornoMensal.iloc[-1, 1]
DiarioBitcoin = retornoDiario.iloc[-1, 1]

AnualIbovespa = retornoAnual.iloc[-1, 2]
MensalIbovespa = retornoMensal.iloc[-1, 2]
DiarioIbovespa = retornoDiario.iloc[-1, 2]

AnualSP500 = retornoAnual.iloc[-1, 3]
MensalSP500 = retornoMensal.iloc[-1, 3]
DiarioSP500 = retornoDiario.iloc[-1, 3]
```

```python
AnualDolar = round(AnualDolar * 100, 2)
MensalDolar = round(MensalDolar * 100, 2)
DiarioDolar = round(DiarioDolar * 100, 2)

AnualBitcoin = round(AnualBitcoin * 100, 2)
MensalBitcoin = round(MensalBitcoin * 100, 2)
DiarioBitcoin = round(DiarioBitcoin * 100, 2)

AnualIbovespa = round(AnualIbovespa * 100, 2)
MensalIbovespa = round(MensalIbovespa * 100, 2)
DiarioIbovespa = round(DiarioIbovespa * 100, 2)

AnualSP500 = round(AnualSP500 * 100, 2)
MensalSP500 = round(MensalSP500 * 100, 2)
DiarioSP500 = round(DiarioSP500 * 100, 2)
```

```python
plt.style.use("cyberpunk")

dadosFechamento.plot(y = 'Dolar', use_index = True, legend = False)
plt.title("Dolar")
plt.savefig('doleta.png', dpi = 300)
plt.show()
```

```python
plt.style.use("cyberpunk")

dadosFechamento.plot(y = 'Bitcoin', use_index = True, legend = False)
plt.title("Bitcoin")
plt.savefig('bitcoio.png', dpi = 300)
plt.show()
```

```python
plt.style.use("cyberpunk")

dadosFechamento.plot(y = 'Ibovespa', use_index = True, legend = False)
plt.title("Ibovespa")
plt.savefig('ibovespa.png', dpi = 300)
plt.show()
```

```python
plt.style.use("cyberpunk")

dadosFechamento.plot(y = 'S&P500', use_index = True, legend = False)
plt.title("S&P500")
plt.savefig('sp500.png', dpi = 300)
plt.show()
```
